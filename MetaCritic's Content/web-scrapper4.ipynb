```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
all_games = []

for page_num in range(1,19):
    url = f"https://www.metacritic.com/browse/game/all/all/current-year/userscore/?page={page_num}"
    headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36', 'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8', 'Accept-Language': 'en-US,en;q=0.5', 'Connection': 'keep-alive' }
    page  = requests.get(url,headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')

    headings = soup.findAll("h3",attrs={"class":"c-finderProductCard_titleHeading"})
    for heading in headings:
        all_games.append(heading.text)

    cleaned_dl_games = [item.split('. ', 1)[1] for item in all_games]

print(len(cleaned_dl_games))
```

```python
df_games = pd.DataFrame(cleaned_dl_games,columns=["GAMES AVAILABLE"])
df_games.head()
```

```python
all_movies = []
for page_num in range(1,687):
    url = f"https://www.metacritic.com/browse/movie/?releaseYearMin=1910&releaseYearMax=2024&page={page_num}"
    headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36', 'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8', 'Accept-Language': 'en-US,en;q=0.5', 'Connection': 'keep-alive' }
    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')

    movie_names = soup.findAll("h3",attrs={"class":"c-finderProductCard_titleHeading"})
    for i in movie_names:
        all_movies.append(i.text)

    cleaned_dl_movies = [item.split('. ', 1)[1] for item in all_movies]
print(cleaned_dl_movies)
```

```python
print(len(cleaned_dl_movies)) #16,447
```

```python
df_movies = pd.DataFrame(cleaned_dl_movies,columns=["Movie Names"])
df_movies.head()
```

```python
all_tv = []
for page_num in range(1,133):
    url = f"https://www.metacritic.com/browse/tv/?releaseYearMin=1910&releaseYearMax=2024&page={page_num}"
    headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36', 'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8', 'Accept-Language': 'en-US,en;q=0.5', 'Connection': 'keep-alive' }
    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')

    tv_names = soup.findAll("h3",attrs={"class":"c-finderProductCard_titleHeading"})
    for i in tv_names:
        all_tv.append(i.text)

    cleaned_dl_tv = [item.split('. ', 1)[1] for item in all_tv]
print(cleaned_dl_tv)
```

```python
df_tv = pd.DataFrame(cleaned_dl_tv,columns=["TV Series"])
df_tv.head()
```

```python
data={
    "GAMES AVAILABLE": cleaned_dl_games,
    "MOVIES MOST LOVED BY PEOPLE": cleaned_dl_movies,
    "TV SHOWS" : cleaned_dl_tv
}

df = pd.DataFrame.from_dict(data, orient='index')

df.head()
```
